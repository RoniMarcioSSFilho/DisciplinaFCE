```python
par=Aquecimento(
    q=5000.0,
    m= 50.0,
    Tamb= 25.0
)

#constante de tempo
tau=(par.m*par.Cp)/(par.U*par.A)

#Valor final da temperatura
Tss = par.Tamb + (par.q/(par.U*par.A))

#Vetor de tempo
tempo=np.linspace(0,5*tau,200)

# Simulação
t,T= simula_aquecimento(tempo, T0=25.0, params=par)

#Gráfico
plt.plot(t, T, label=rf'$\tau$ = {tau:.2f} s')
plt.axhline(Tss, linestyle='--', color='red',
            label=rf'$T_{{ss}}$ = {Tss:.2f} °C')

plt.xlabel("Tempo [s]")
plt.ylabel("Temperatura [°C]")
plt.title("Resposta dinâmica do tanque aquecido")
plt.legend()

plt.grid(True)
plt.tight_layout()

plt.show()
```
