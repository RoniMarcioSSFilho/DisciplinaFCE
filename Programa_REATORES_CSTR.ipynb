```python
from dataclasses import dataclass
import numpy as np

@dataclass
class CSTR:
    """
    Classe que representa um reator CSTR.

    Atributos
    ---------
    F : float
        Vazão volumétrica do fluido (L/s ou m³/s).

    V : float
        Volume do reator (L ou m³).

    k : float
        Constante cinética da reação (1/s).
    """
    F: float  # vazão
    V: float  # volume
    k: float  # constante cinética

def modelo_cstr(CAin, reator):
    """
    Calcula a concentração de saída de um CSTR.

    Parâmetros
    ----------
    CAin : float
        Concentração de entrada no reator.

    reator : CSTR
        Objeto contendo os parâmetros do reator.

    Retorno
    -------
    CAout : float
        Concentração de saída do reator.

    Equação
    --------
    Para reação de primeira ordem em regime estacionário:

    CAout = (F / (F + V*k)) * CAin
    """

    CAout = (reator.F / (reator.F + reator.V * reator.k)) * CAin

    return CAout

def simula_cstr(num_estagios,reator):
    """
    Simula vários CSTRs em série.

    Parâmetros
    ----------
    num_estagios : int
        Número de reatores em série.

    reator : CSTR
        Objeto contendo os parâmetros do reator.

    Retorno
    -------
    CA : list
        Lista com as concentrações em cada estágio.

    X : list
        Lista com as conversões em cada estágio.

    Observação
    ----------
    A função utiliza:
    - CA0 : concentração inicial
    - F, V, k : parâmetros do reator

    Esses valores devem estar definidos anteriormente no código.
    """
    CA = [CA0]

    # Simulação dos reatores
    for i in range(num_estagios):

        reator = CSTR(F, V, k)
        CA_saida = modelo_cstr(CA[-1], reator)
        CA.append(CA_saida)

    # Conversões
    X = []

    for c in CA[1:]:
        conversao = (CA0 - c) / CA0
        X.append(conversao)

    return CA, X

```
