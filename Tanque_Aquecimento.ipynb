```python
from dataclasses import dataclass
import numpy as np
from scipy.integrate import solve_ivp


@dataclass
class Aquecimento:

  U: float = 150.0
  A: float = 5.0
  Tamb: float = 25.0
  q: float = 5000.0
  m: float = 50.0
  Cp: float = 1670.0


def modelo_aquecimento(t, T, params):

  U = params.U
  A = params.A
  Tamb = params.Tamb
  q = params.q
  m = params.m
  Cp = params.Cp

  dTdt = (U * A * (Tamb - T) + q) / (m * Cp)

  return dTdt


def simula_aquecimento(tempo, T0=25.0, params=None):

    if params is None:
        params = Aquecimento()

    sol = solve_ivp(
        modelo_aquecimento,
        t_span=(tempo[0], tempo[-1]),
        y0=[T0],
        t_eval=tempo,
        args=(params,),
        method="BDF"
    )

    return sol.t, sol.y[0]

```
