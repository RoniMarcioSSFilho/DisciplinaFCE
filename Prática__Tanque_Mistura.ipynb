```python
from dataclasses import dataclass

@dataclass
class Corrente:
  """
  Vazão: vazão da corrente em kg/h
  Fração: fração mássica kg/kg
  """
  vazao: float
  fracao: float
```

```python
def misturador(c1: Corrente, c2: Corrente):
  """
  Calcula a vazão e a fraçã mássica de uma mistura de correntes.
  Entrada:
    c1: Corrente
    c2: Corrente
  Saída:
    F: Vazão total
    x: Fração mássica total
  """

  Fa= c1.vazao
  xa= c1.fracao

  Fb= c2.vazao
  xb= c2.fracao

  F= Fa + Fb

  x=(Fa*xa+Fb*xb)/F

  return F,x
```
